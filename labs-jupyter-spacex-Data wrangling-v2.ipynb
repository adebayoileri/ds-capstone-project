```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv")
df.head(10)
```

```python
# Identify and calculate the percentage of missing values in each attribute
df.isnull().sum() / len(df) * 100
```

```python
# Identify which columns are numerical and categorical
df.dtypes
```

### TASK 1: Calculate the number of launches on each site

```python
# TASK 1: Apply value_counts() on column LaunchSite
df['LaunchSite'].value_counts()
```

### TASK 2: Calculate the number and occurrence of each orbit

```python
# TASK 2: Apply value_counts on Orbit column
df['Orbit'].value_counts()
```

### TASK 3: Calculate the number and occurrence of mission outcome of the orbits

```python
# TASK 3: Apply value_counts() on Outcome column
landing_outcomes = df['Outcome'].value_counts()
landing_outcomes
```

```python
# Print each outcome with its index
for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)
```

```python
# Create a set of bad outcomes (unsuccessful landings)
bad_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]])
bad_outcomes
```

### TASK 4: Create a landing outcome label from Outcome column

```python
# TASK 4: Create landing_class
# landing_class = 0 if bad_outcome, 1 otherwise
landing_class = [0 if outcome in bad_outcomes else 1 for outcome in df['Outcome']]
landing_class[:10]
```

```python
# Assign to Class column
df['Class'] = landing_class
df[['Class']].head(8)
```
